```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_results #, unique
import matplotlib.cm as cm
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
import trimap

cols = cm.tab10(np.linspace(0, 1, 10))
```

```python
data = pd.read_csv('./data/OnlineNewsPopularity.csv')
```

```python
data.columns
```

```python
X = data.loc[:,' timedelta':' shares'].values
```

```python
y_pca = PCA(n_components = 2).fit_transform(X)
y_tsne = TSNE().fit_transform(X)
y_umap = umap.UMAP().fit_transform(X)
y_trimap = trimap.TRIMAP(verbose=False).fit_transform(X)
```

## Clusters in high and low dimension

We color the points based on the cluster indices in high-dimension obtained using k-mean algorithm

```python
L = KMeans(n_clusters = 2).fit(X).labels_
plot_results([y_tsne,y_umap,y_trimap,y_pca], cols[L,:], extension='(k=2)')
```

```python
L = KMeans(n_clusters = 3).fit(X).labels_
plot_results([y_tsne,y_umap,y_trimap,y_pca], cols[L,:], extension='(k=3)')
```

```python
L = KMeans(n_clusters = 5).fit(X).labels_
plot_results([y_tsne,y_umap,y_trimap,y_pca], cols[L,:], extension='(k=5)')
```

```python
L = KMeans(n_clusters = 10).fit(X).labels_
plot_results([y_tsne,y_umap,y_trimap,y_pca], cols[L,:], extension='(k=10)')
```
